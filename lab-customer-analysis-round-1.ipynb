```python
#rearranging columns in dataframe2
file2 = file2.loc[:,['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
#updating the column_names_2 with the new column orders
column_names_2 = file2.columns
#checking if the columns are rearranged
#print(column_names_2)
```

```python
#concatenating all the df vertically in a df called 'data'
data = pd.concat([file1, file2, file3], axis=0)
#checking if we have all the rows and if the number of columns remain the same 
data.shape
data
```

```python
#standardizing columns name
column_name = []  #defining an empty list to store the data column names
for i in data.columns: #for loop in each column of the df data
    i = i.lower() #each column name is changed to lowercase 
    if (' ' in i): #condition to check if there's a blank space in the column name
        i = i.replace(' ','_')    #replace ' ' by '_'
    column_name.append(i) #add in the list each column name in each loop iteration 
data.columns = column_name #attributing the list created to the data columns
print(data.columns)
```

```python
#cleaning data
#deleting the columns education and open complaints
data_cleaned = data.copy() #creating a new dataframe to start cleaning data
data_cleaned = data_cleaned.drop([('education'),('number_of_open_complaints')],axis=1) #deleting columns

data_cleaned
```

```python
#cleaning data
#removing '%' symbol and dividing by 100
def clean_symbol(x): #creating a new function to replace "%" by ""
    x = str(x) #transforming x in string
    if ( "%" in x ):
        x = x.replace("%","")
    x = float(x) #transforming x in float
    x = (x/100.0) 
    return x
data_cleaned['customer_lifetime_value'] = data_cleaned['customer_lifetime_value'].apply(clean_symbol) #applying function in each value of the column

data_cleaned[['customer_lifetime_value']]

data_cleaned.dtypes
#it shows that customer_lifetime_value is float64
```

```python
#cleaning data
#removing duplicate rows

data_cleaned = data_cleaned.drop_duplicates() #remove all the duplicate rows

data_cleaned.shape #shows that now we have 2065 rows without the duplicated rows
```
